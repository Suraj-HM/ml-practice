```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Position_Salaries.csv')
X = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(X, y)
```

```python
dtr.predict([[6.5]])
```

```python
plt.scatter(X, y)
plt.plot(X, dtr.predict(X), color='orange')
plt.show()
```

```python
# High resolution
x_n = np.arange(min(X), max(X), 0.1)
x_n = x_n.reshape(len(x_n), 1)
plt.scatter(X, y)
plt.plot(x_n, dtr.predict(x_n), color='orange')
plt.show()
```
